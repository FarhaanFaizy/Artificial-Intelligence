```python
def dfs(d,source,key):
    open_list=[source]
    closed_list=[]
    while open_list:
        source=open_list.pop(0)
        closed_list.append(source)
        if source==key:
            return [closed_list,open_list,1]
        for i in d[source]:
            if i not in closed_list and i not in open_list:
                open_list.insert(0,i)
    return [closed_list,open_list,0]
def depthxtree(graph,depth,source):
    c=1
    d={}
    templis=[]
    all_nodes=[]
    all_nodes.append(source+str(c))
    while c<=depth:
        if c==1:
            d[source]=[]
            templis=graph[source]
            parent=source
        if c==2:
            for i in templis:
                d[i]=[]
                d[parent].append(i)
                all_nodes.append(i+str(c))
        else:
            for j in all_nodes:
                if j[-1]==str(c-1):
                    templis=graph[j[:-1]]
                    for i in templis:
                        d[i]=[]
                        d[j[:-1]].append(i)
                        all_nodes.append(i+str(c))
        c+=1
    return d
                
def iddfs(graph,source,key):
    depth=1
    while 1:
        d=depthxtree(graph,depth,source)
        y=dfs(d,source,key)
        print("DEPTH=",depth)
        print("Open_list=",y[1])
        print("Closed_list=",y[0])
        depth+=1
        if len(d)==len(graph) or y[2]==1:
            break
n=int(input("Enter number of nodes in graph:"))
for i in range(n):
    print(chr(65+i),end=' ')
print("are the nodes")
graph={}
for i in range(n):
    print("Enter the child nodes of",chr(i+65),end='')
    c=list(map(str,input(":").split()))
    if c[0]!='0':
        graph[chr(i+65)]=c
    else:
        graph[chr(i+65)]=[]
# graph={'A':['B','C','D'],'B':['E','F'],'C':['G','H'],'D':['I','J'],'E':['K','L'],'F':['L','M'],
#   'G':['N'],'H':['O','P'],'I':['P','Q'],'J':['R'],'K':['S'],'L':['T'],'P':['U'],'M':[],
#    'N':[],'O':[],'Q':[],'R':[],'S':[],'T':[],'U':[]}
iddfs(graph,'A','U')
```
